```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
gdp_df = pd.read_csv('../data/gdp_percapita.csv', nrows = 6606)
gdp_df = gdp_df.drop(columns='Value Footnotes')
gdp_df.columns=['Country', 'Year', 'GDP_Per_Capita']
gdp_df
```

```python
internet_df = pd.read_csv('../data/internet_use.csv', nrows=4495)
internet_df = internet_df.drop(columns='Value Footnotes')
internet_df.columns=['Country', 'Year', 'Internet_Usage']
internet_df
```

```python
gdp_and_internet_use = pd.merge(internet_df, gdp_df,
                               left_on = ['Country','Year'], right_on = ['Country','Year'],
                               how = 'outer')
gdp_and_internet_use
```

```python
subset = gdp_and_internet_use.query('Year == 2004 or Year == 2009 or Year == 2014')
subset
```

```python
yr04 = subset.query('Year == 2004')
yr04

yr09 = subset.query('Year == 2009')
yr09

yr14 = subset.query('Year == 2014')
yr14
```

```python
yr14max = yr14.sort_values('Internet_Usage', ascending= False)
print(yr14max)
yr14min = yr14.sort_values('Internet_Usage', ascending= True)
print(yr14min)
```

```python
yr04max = yr04.sort_values('Internet_Usage', ascending= False)
print(yr04max)
yr04min = yr04.sort_values('Internet_Usage', ascending= True)
print(yr04min)
```

```python
yr09max = yr09.sort_values('Internet_Usage', ascending= False)
print(yr09max)
yr09min = yr09.sort_values('Internet_Usage', ascending= True)
print(yr09min)
```

```python
plt.scatter(x = 'Internet_Usage', y = 'GDP_Per_Capita', data = yr04, alpha = .1)
plt.title("2004")
```

```python
plt.scatter(x = 'Internet_Usage', y = 'GDP_Per_Capita', data = yr09, alpha = .1)
plt.title("2009")
```

```python
plt.scatter(x = 'Internet_Usage', y = 'GDP_Per_Capita', data = yr14, alpha = .1)
plt.title("2014")
```

<h3>You can tell that as the years increase, the poorer countries are slowly climbing in internet usage. The GDP is not Unimodal in 2014, but the internet usage is Unimodal in 2014.

```python
yr14max = yr14.sort_values('Internet_Usage', ascending= False)
yr14max.head()
```

The Top 5 Internet Usage Countries in 2014 are:
    <li>Iceland</li>
    <li>Bermuda</li>
    <li>Norway</li>
    <li>Denmark</li>
    <li>Andorra</li>
        

```python
top_countries = ['Iceland', 'Bermuda', 'Norway', 'Denmark', 'Andorra']
years = ['2004', '2009', '2014']

top_5_internet = gdp_and_internet_use.query(('Country in @top_countries & Year in @years'))

top_5_internet
```

```python
IntUseOverTime = sns.FacetGrid(top_5_internet, hue = 'Country', height = 5)
IntUseOverTime.map(plt.plot, 'Year', 'Internet_Usage', alpha = .5, marker='o').set_titles('Internet Use per Year')
IntUseOverTime.set(xticks = [2004,2009,2014])
IntUseOverTime.add_legend()
```
